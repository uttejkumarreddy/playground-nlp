```python
import re

import nltk
# nltk.download('stopwords')
# nltk.download('punkt')
# nltk.download('words')
# nltk.download('wordnet')
# nltk.download('omw-1.4')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
# from nltk.tokenize import word_tokenize

words = list(nltk.corpus.words.words())
stopwords = list(stopwords.words('english'))

lemmatizer = WordNetLemmatizer()

def cleanTweets(tweet):
  tweet = tweet.lower()
  tweet = re.sub(r'@[A-Za-z0-9]+', '', tweet) # Remove mentions
  tweet = re.sub(r'#', '', tweet) # Remove hashtags
  tweet = re.sub(r'https?://\S+', '', tweet) # Remove urls
  tweet = bytes(tweet, 'utf-8').decode('utf-8', 'ignore') # Remove non utf-8 characters

  # Lemmatize tweet
  tweet = ' '.join([lemmatizer.lemmatize(word) for word in tweet.split()])

  # Remove stopwords
  tweet = ' '.join([word for word in str(tweet).split() if word not in stopwords])
  
  # Remove emoticons
  emoticon_pattern = re.compile(u'(' + u'|'.join(k for k in emoticons) + u')')
  tweet = emoticon_pattern.sub(r'', tweet)

  return tweet

df['Tweets'] = df['Tweets'].apply(cleanTweets)
```
