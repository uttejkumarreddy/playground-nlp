```python
from haystack.nodes import TransformersSummarizer
from haystack import Document

import pandas as pd
import evaluate

from rouge import Rouge
```

<h1> Data </h1>

```python
df = pd.read_csv('aesopfables-summaries.csv', encoding = 'latin-1')

originalText   = df['text'].tolist()
summarizedText = df['summary'].tolist()
```

<h1> PEGASUS </h1>
https://huggingface.co/google/pegasus-xsum </br>
Trained on C4 and HugeNews

```python
summarizer = TransformersSummarizer(model_name_or_path="google/pegasus-xsum", max_length = 200)
```

```python
predictedSummaries = []
for i in range(len(originalText)):
  doc = [Document(originalText[i])]
  predictedSummary = summarizer.predict(documents=doc)[0].meta['summary']
  predictedSummaries.append(predictedSummary)
```

```python
rouge = evaluate.load('rouge')
rouge.compute(predictions = predictedSummaries, references = summarizedText)
```
