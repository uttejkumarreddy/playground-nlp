```python
#Importing Libraries
import torch
from transformers import BertForQuestionAnswering
from transformers import BertTokenizer
import numpy as np
import pandas as pd
import torch
from transformers import BertForQuestionAnswering
from transformers import BertTokenizer
from transformers import AutoTokenizer
from transformers import pipeline
```

```python
#Loading the datset
df = pd.read_csv("/content/bedtimestories.csv", encoding= 'latin1')
```

```python
df.head()
```

```python
#Loading the annotated data done using haystack annotation tool
new_df = pd.read_csv('/content/answers.csv')
new_df
```

```python
# picking out a random question and answer pair from the dataset
import numpy as np
random_num = np.random.randint(0,len(new_df))
question = new_df["question"][random_num]
text = new_df["context"][random_num]
answer = new_df["text"][random_num]
```

```python
from transformers import pipeline
#Importing the moodel using Hugging face pipeline for inference
model_id1 = "deepset/bert-medium-squad2-distilled"
question_answerer_1 = pipeline("question-answering", model=model_id1)
```

```python
result = question_answerer_1(question= question, context=text)
print("Question : ", question)
print("Context : " , text)
print("Actual answer : ", answer)
print(
    f"Pred Answer: '{result['answer']}', score: {round(result['score'], 4)}, start: {result['start']}, end: {result['end']}"
)
```

```python
model_id2 = "bert-large-uncased-whole-word-masking-finetuned-squad"
question_answerer_2 = pipeline("question-answering", model=model_id2)
```

```python
result = question_answerer_2(question= question, context=text)
print("Question : ", question)
print("Context : " , text)
print("Actual answer : ", answer)
print(
    f"Pred Answer: '{result['answer']}', score: {round(result['score'], 4)}, start: {result['start']}, end: {result['end']}"
)
```

```python
model_id3 = "deepset/bert-base-cased-squad2"
question_answerer_3 = pipeline("question-answering", model=model_id3)
```

```python
result = question_answerer_3(question= question, context=text)
print("Question : ", question)
print("Context : " , text)
print("Actual answer : ", answer)
print(
    f"Pred Answer: '{result['answer']}', score: {round(result['score'], 4)}, start: {result['start']}, end: {result['end']}"
)
```

```python
#Calculating F1-Score and Exact Match for 50 Predicted and Exact Answers Q/A pairs using model3
predicted_answers = []
actual_answers = []
i = 0
for _,row in new_df.iterrows():
  question = row["question"]
  context = row["context"]
  id = str(row["answer_id"])
  answer = row["text"]
  answer_start = row["answer_start"]
  actual_answers.append({"id" : id, 'answers': {'text': [answer], 'answer_start': [answer_start]}})
  result = question_answerer_3(question= question, context=context)
  predicted_answers.append({"id": id, "prediction_text" : result['answer']})
  i += 1
  if i== 50:
    break
```

```python
#Using Hugging Face inference to evaluate metrics such as F1-Score and Exact Match for first 50 Q/A pairs
import evaluate
metric = evaluate.load("squad")
metric.compute(predictions=predicted_answers, references=actual_answers)
```

```python
#Printing the Predicted answer and accurate answer for 50 Q/A pairs
for item1, item2 in zip(predicted_answers, actual_answers):
  print(item1, item2)
```
